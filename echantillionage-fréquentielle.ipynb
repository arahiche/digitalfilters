**Échantillonnage Fréquentiel**

Cette méthode repose sur l'idée d'échantillonner la réponse en fréquence d'un filtre idéal pour obtenir le filtre numérique

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition des paramètres
wc = 0.5 * np.pi  # Fréquence de coupure
omega = np.linspace(-np.pi, np.pi, 1000)  # Fréquence analogique

# Réponse en fréquence du filtre passe-bas
Ha = np.where(np.abs(omega) <= wc, 1, 0)

# Plot de la réponse en fréquence
plt.figure(figsize=(8, 4))
plt.plot(omega, Ha)
plt.title("Réponse en Fréquence d'un Filtre Passe-Bas Idéal")
plt.xlabel("Fréquence ω (rad/s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

Transformation vers le Filtre Numérique et Échantillonnage

On échantillonne la réponse fréquentielle analogique pour en faire un filtre numérique :

```python
# Nombre de points d'échantillonnage
N = 64
k = np.arange(N)
omega_k = 2 * np.pi * k / N  # Points d'échantillonnage

# Réponse en fréquence du filtre numérique
H = np.where(np.abs(omega_k) <= wc, 1, 0)

# Plot de la réponse échantillonnée
plt.figure(figsize=(8, 4))
plt.stem(omega_k - np.pi, H, basefmt=" ")
plt.title("Réponse en Fréquence Échantillonnée du Filtre")
plt.xlabel("Fréquence ω (rad/s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

Calcul de la Réponse Impulsionnelle via IDFT

On calcule la réponse impulsionnelle h[n] du filtre numérique en utilisant l'IDFT des coefficients échantillonnés.

```python
# Calcul de la réponse impulsionnelle
h = np.fft.ifft(H, N).real

# Plot de la réponse impulsionnelle
plt.figure(figsize=(8, 4))
plt.stem(h, basefmt=" ")
plt.title("Réponse Impulsionnelle du Filtre Passe-Bas Numérique")
plt.xlabel("n")
plt.ylabel("Amplitude h[n]")
plt.grid()
plt.show()
```
